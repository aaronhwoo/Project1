```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("MasterDataframe.csv")
df.head(50)
```

```python
outdoor_df = df.loc[df["Indoor"] == False, :]
outdoor_df.head()
```

```python
outdoor_df["Total Points"] = outdoor_df["Total Points"].astype(int)
outdoor_df["Temp"] = outdoor_df["Temp"].astype(float)
```

```python
plt.scatter(outdoor_df["Temp"], outdoor_df["Total Points"], c= "lightblue",
            edgecolor="black", linewidths=1, marker="o", alpha = 0.6)

plt.title(f"Gametime Temperature vs. Total Points")
plt.ylabel("Total Points")
plt.xlabel("Gametime Temperature (F)")
plt.grid(True)
plt.xlim([(outdoor_df["Temp"].min()-5),(outdoor_df["Temp"].max()+5)])
plt.ylim([(outdoor_df["Total Points"].min()-5),(outdoor_df["Total Points"].max()+5)])

plt.savefig("Temps_Points.png")
plt.show()
```

```python
indoor_df["Home FG Made 1-19"] = indoor_df["Home FG Made 1-19"].astype(int)
indoor_df["Home FG Made 20-29"] = indoor_df["Home FG Made 20-29"].astype(int)
indoor_df["Home FG Made 30-39"] = indoor_df["Home FG Made 30-39"].astype(int)
indoor_df["Home FG Made 40-49"] = indoor_df["Home FG Made 40-49"].astype(int)
indoor_df["Home FG Made 50+"] = indoor_df["Home FG Made 50+"].astype(int)
indoor_df["Away FG Made 1-19"] = indoor_df["Away FG Made 1-19"].astype(int)
indoor_df["Away FG Made 20-29"] = indoor_df["Away FG Made 20-29"].astype(int)
indoor_df["Away FG Made 30-39"] = indoor_df["Away FG Made 30-39"].astype(int)
indoor_df["Away FG Made 40-49"] = indoor_df["Away FG Made 40-49"].astype(int)
indoor_df["Away FG Made 50+"] = indoor_df["Away FG Made 50+"].astype(int)

outdoor_df["Home FG Made 1-19"] = outdoor_df["Home FG Made 1-19"].astype(int)
outdoor_df["Home FG Made 20-29"] = outdoor_df["Home FG Made 20-29"].astype(int)
outdoor_df["Home FG Made 30-39"] = outdoor_df["Home FG Made 30-39"].astype(int)
outdoor_df["Home FG Made 40-49"] = outdoor_df["Home FG Made 40-49"].astype(int)
outdoor_df["Home FG Made 50+"] = outdoor_df["Home FG Made 50+"].astype(int)
outdoor_df["Away FG Made 1-19"] = outdoor_df["Away FG Made 1-19"].astype(int)
outdoor_df["Away FG Made 20-29"] = outdoor_df["Away FG Made 20-29"].astype(int)
outdoor_df["Away FG Made 30-39"] = outdoor_df["Away FG Made 30-39"].astype(int)
outdoor_df["Away FG Made 40-49"] = outdoor_df["Away FG Made 40-49"].astype(int)
outdoor_df["Away FG Made 50+"] = outdoor_df["Away FG Made 50+"].astype(int)
```

```python
indoor_fg_made_1_19 = (indoor_df["Home FG Made 1-19"].sum()) + (indoor_df["Away FG Made 1-19"].sum())
indoor_fg_made_20_29 = (indoor_df["Home FG Made 20-29"].sum()) + (indoor_df["Away FG Made 20-29"].sum())
indoor_fg_made_30_39 = (indoor_df["Home FG Made 30-39"].sum()) + (indoor_df["Away FG Made 30-39"].sum())
indoor_fg_made_40_49 = (indoor_df["Home FG Made 40-49"].sum()) + (indoor_df["Away FG Made 40-49"].sum())
indoor_fg_made_50 = (indoor_df["Home FG Made 50+"].sum()) + (indoor_df["Away FG Made 50+"].sum())

indoor_fg_att_1_19 = (indoor_df["Home FG Att 1-19"].sum()) + (indoor_df["Away FG Att 1-19"].sum())
indoor_fg_att_20_29 = (indoor_df["Home FG Att 20-29"].sum()) + (indoor_df["Away FG Att 20-29"].sum())
indoor_fg_att_30_39 = (indoor_df["Home FG Att 30-39"].sum()) + (indoor_df["Away FG Att 30-39"].sum())
indoor_fg_att_40_49 = (indoor_df["Home FG Att 40-49"].sum()) + (indoor_df["Away FG Att 40-49"].sum())
indoor_fg_att_50 = (indoor_df["Home FG Att 50+"].sum()) + (indoor_df["Away FG Att 50+"].sum())

indoor_fg_miss_1_19 = indoor_fg_att_1_19 - indoor_fg_made_1_19
indoor_fg_miss_20_29 = indoor_fg_att_20_29 - indoor_fg_made_20_29
indoor_fg_miss_30_39 = indoor_fg_att_30_39 - indoor_fg_made_30_39
indoor_fg_miss_40_49 = indoor_fg_att_40_49 - indoor_fg_made_40_49
indoor_fg_miss_50 = indoor_fg_att_50 - indoor_fg_made_50

mades = [indoor_fg_made_1_19, indoor_fg_made_20_29, indoor_fg_made_30_39, indoor_fg_made_40_49, indoor_fg_made_50]
misses = [indoor_fg_miss_1_19, indoor_fg_miss_20_29, indoor_fg_miss_30_39, indoor_fg_miss_40_49, indoor_fg_miss_50]

N = 5
ind = np.arange(N)    # the x locations for the groups
width = 0.5       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, mades, width, color = "g")
p2 = plt.bar(ind, misses, width, color = "r",
             bottom=mades)

plt.ylabel('Field Goals')
plt.xlabel('Field Goal Distance (Yds)')
plt.title('Indoor Kicking Results')
plt.xticks(ind, ("0-19", "20-29", "30-39", '40-49', '50+'))
plt.yticks(np.arange(0, 50, 10))
plt.legend((p1[0], p2[0]), ('Made', 'Missed'))

plt.savefig("Indoor_Kicking_Results.png", bbox_inches = "tight")
plt.show()
```
