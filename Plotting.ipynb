```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("MasterDataframe.csv")
df.head(50)
```

```python
outdoor_df = df.loc[df["Indoor"] == False, :]
outdoor_df = outdoor_df.loc[outdoor_df["Temp"] != "No Data", :]
outdoor_df.head()
```

```python
outdoor_df["Total Points"] = outdoor_df["Total Points"].astype(int)
outdoor_df["Temp"] = outdoor_df["Temp"].astype(float)
```

```python
plt.scatter(outdoor_df["Temp"], outdoor_df["Total Points"], c= "lightblue",
            edgecolor="black", linewidths=1, marker="o", alpha = 0.6)

plt.title(f"Gametime Temperature vs. Total Points")
plt.ylabel("Total Points")
plt.xlabel("Gametime Temperature (F)")
plt.grid(True)
plt.xlim([(outdoor_df["Temp"].min()-5),(outdoor_df["Temp"].max()+5)])
plt.ylim([(outdoor_df["Total Points"].min()-5),(outdoor_df["Total Points"].max()+5)])

plt.savefig("Temps_Points.png")
plt.show()
```
