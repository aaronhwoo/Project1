```python
import pandas as pd
```

```python
games_df = pd.read_csv("GameResults.csv")
games_df.head(32)
```

```python
stadiums_df = pd.read_csv("Stadiums.csv")
stadiums_df.head(32)
```

```python
merged_df = pd.merge(games_df, stadiums_df, how="left", on="Location")
merged_df.head(32)
```

```python
merged_df.to_csv("GameResultsLatsLongs.csv", index=False, header=True)
```
