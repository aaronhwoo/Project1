```python
import base64
import requests
from pprint import pprint
import json
import pandas as pd
import time
```

```python
weeks = []
gameids = []
gameurls = []
dates = []
compactdates = []
gametimes = []
locations = []
hometeams = []
awayteams = []
homescores = []
homerushyards = []
homepassyards = []
hometotalyards = []
homefirstdowns = []
homerushattempts = []
homerushtds = []
homepassattempts = []
homepasscompletions = []
homepasstds = []
homefgattempts = []
homefgmades = []
homepunts = []
homefums = []
homeints = []
awayscores = []
awayrushyards = []
awaypassyards = []
awaytotalyards = []
awayfirstdowns = []
awayrushattempts = []
awayrushtds = []
awaypassattempts = []
awaypasscompletions = []
awaypasstds = []
awayfgattempts = []
awayfgmades = []
awaypunts = []
awayfums = []
awayints = []

response = requests.get(
                url="https://api.mysportsfeeds.com/v1.2/pull/nfl/2018-regular/full_game_schedule.json",
                headers={
                    "Authorization": "Basic " + base64.b64encode('{}:{}'.format("caa67d28-e6d3-4fae-8919-10331e","password").encode('utf-8')).decode('ascii')
                }).json()
#pprint(response)

for x in range(0,256):
    week = response["fullgameschedule"]["gameentry"][x]["week"]
    gameid = response["fullgameschedule"]["gameentry"][x]["id"]
    date = response["fullgameschedule"]["gameentry"][x]["date"]
    gametime = response["fullgameschedule"]["gameentry"][x]["time"]
    location = response["fullgameschedule"]["gameentry"][x]["location"]
    hometeam = response["fullgameschedule"]["gameentry"][x]["homeTeam"]["Abbreviation"]
    awayteam = response["fullgameschedule"]["gameentry"][x]["awayTeam"]["Abbreviation"]
    compactdate = date.replace("-", "")
    gameurl = f"{compactdate}-{awayteam}-{hometeam}"
    weeks.append(week)
    gameids.append(gameid)
    gameurls.append(gameurl)
    dates.append(date)
    compactdates.append(compactdate)
    gametimes.append(gametime)
    locations.append(location)
    hometeams.append(hometeam)
    awayteams.append(awayteam)    
    
games_df = pd.DataFrame({"Week": weeks, 
                        "Game ID": gameids, 
                        "Game URL": gameurls,
                        "Date": dates,
                        "Compact Date": compactdates,
                        "Time": gametimes, 
                        "Location": locations, 
                        "Home Team": hometeams,
                        "Away Team": awayteams})

games_df.head(25)
```
