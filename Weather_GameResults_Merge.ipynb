```python
import pandas as pd
import numpy as np
import requests
import json
from pprint import pprint
import sys
sys.path.append('..')
from config import darksky_api_key
```

```python
gameresults_df = pd.read_csv("GameResults_test.csv")
gameresults_df.head(30)
```

```python
gameresults_df['Summary'] = ""
gameresults_df['Wind Speed'] = ""
gameresults_df['Wind Gust'] = ""
gameresults_df['Temperature'] = ""
gameresults_df.head(30)
```

```python
base_url = "https://api.darksky.net/forecast/"

params = {"key":darksky_api_key,
         }

for index, row in gameresults_df.iterrows():
    #base_url = "https://api.darksky.net/forecast/"
    
    lat = row['Lat']
    lng = row['Lng']
    time = row['Unix Time']
    
# I excluded hourly and flag requests from DarkskyAPI just so that I can view the data easily
    response = requests.get(base_url + str(darksky_api_key) + '/' + str(lat) + ',' + str(lng) + ',' + str(time)+ '?' + 'exclude=hourly,flags')
    
    #print(response.url)
    #response = requests.get(base_url, params=params).json()
    
    data = response.json()
    
    #pprint (response)
    try:
        gameresults_df.loc[index, "Summary"] = data["currently"]["summary"]
        gameresults_df.loc[index, "Wind Speed"] = data["currently"]["windSpeed"]
        gameresults_df.loc[index, "Wind Gust"] = data["currently"]["windGust"]
        gameresults_df.loc[index, "Temperature"] = data["currently"]["temperature"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```
