```python
import base64
import requests
from pprint import pprint
import json
import pandas as pd
from config import darksky_api_key
import time
```

```python
times_df = pd.read_csv("UnixTimes.csv")
times_df.head(32)
```

```python
weatherurls =[]
```

```python
for index, row in times_df.iterrows():
    base_url = "https://api.darksky.net/forecast/"
    gamelat = row["Lat"]
    gamelng = row["Lng"]
    gametime = row["Unix Time"]
    weatherurl = f"{base_url}{darksky_api_key}/{gamelat},{gamelng},{gametime}"
    weatherurls.append(weatherurl)
```

```python
times_df["Darksky URL"] = weatherurls
times_df.head(32)
```

```python
summarys = []
precipintensitys = []
precipprobs = []
preciptypes = []
temps = []
apparenttemps = []
humiditys = []
windspeeds = []
windgusts = []
cloudcovers = []
uvindexs = []
visibilitys = []
```

```python
for index, row in times_df.iterrows():
    time.sleep(.1)
    gameid = row["Game ID"]
    print(f"Now Processing: {gameid}")
    url = row["Darksky URL"]
    weatherresponse = requests.get(url).json()
    summary = weatherresponse["currently"].get("summary", "No Data")
    precipintensity = weatherresponse["currently"].get("precipIntensity", "No Data")
    precipprob = weatherresponse["currently"].get("precipProbability", "No Data")
    preciptype = weatherresponse["currently"].get("precipType", "No Data")
    temp = weatherresponse["currently"].get("temperature", "No Data")
    apparenttemp = weatherresponse["currently"].get("apparentTemperature", "No Data")
    humidity = weatherresponse["currently"].get("humidity", "No Data")
    windspeed = weatherresponse["currently"].get("windSpeed", "No Data")
    windgust = weatherresponse["currently"].get("windGust", "No Data")
    cloudcover = weatherresponse["currently"].get("cloudCover", "No Data")
    uvindex = weatherresponse["currently"].get("uvIndex", "No Data")
    visibility = weatherresponse["currently"].get("visibility", "No Data")
    
    summarys.append(summary)
    precipintensitys.append(precipintensity)
    precipprobs.append(precipprob)
    preciptypes.append(preciptype)
    temps.append(temp)
    apparenttemps.append(apparenttemp)
    humiditys.append(humidity)
    windspeeds.append(windspeed)
    windgusts.append(windgust)
    cloudcovers.append(cloudcover)
    uvindexs.append(uvindex)
    visibilitys.append(visibility)
```

```python
times_df["Summary"] = summarys
times_df["Precip Intensity"] = precipintensitys
times_df["Precip Probability"] = precipprobs
times_df["Precip Type"] = preciptypes
times_df["Temp"] = temps
times_df["Apparent Temp"]= apparenttemps
times_df["Humidity"] = humiditys
times_df["Wind Speed"] = windspeeds
times_df["Wind Gust"] = windgusts
times_df["Cloud Cover"] = cloudcovers
times_df["UV Index"] = uvindexs
times_df["Visibility"] = visibilitys
times_df.head()
```

```python
weather_df = times_df.drop(columns=["Date", "Time (EST)", "Lat", "Lng", "DateTime", "Unix Time", "Darksky URL"])
weather_df.head(100)
```

```python
weather_df.to_csv("WeatherConditions.csv", index=False, header=True)
```
